```python
n = 15
my_list = []
for x in range(1,n+1):
    my_list.append(str(x))
    if (x % 3) == 0:
        my_list[-1] = "Fizz"
    if (x % 5) == 0:
        my_list[-1] = "Buzz"
    if (x % 3) == 0 and (x % 5) == 0:
        my_list[-1] = "FizzBuzz"
my_list = [str(r) for r in my_list]
print(my_list)
```
